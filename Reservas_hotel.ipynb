```python
import pandas as pd

caminho = 'booking.csv'

df = pd.read_csv(caminho)

df
```

```python
#detectar se há valores ausentes e o tipo das variáveis

df.info()
```

```python
# análise visual das variáveis númericas através de um gráfico de frequências (histograma)
df.hist(bins=15, figsize=(15,10));
```

```python
#corrigindo o erro na string após identificar uma data inexistente

old_string = '2018-2-29'
new_string = '02/28/2018'

df['date of reservation'] = df['date of reservation'].replace(old_string, new_string)
```

```python
# Converta a coluna de data para o tipo datetime
df['date of reservation'] = df['date of reservation'].astype('datetime64[ns]')

df['date of reservation']
```

```python
#verificar a relação entre as colunas numéricas do dataframe

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

numeric_columns = df.select_dtypes(include='number')

# Calcula a matriz de correlação
correlation_matrix = numeric_columns.corr()

# Cria o heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Reds', fmt='.2f', linewidths=.5)
plt.title('Heatmap de Correlação entre Colunas Numéricas')
plt.show()
```

```python
# Função para atribuir classes com base nos gastos médios
def atribuir_classe(valor):
    if valor >= 120:
        return 'Classe_A'
    elif valor < 119 and valor >= 99.45:
        return 'Classe_B'
    elif valor < 99.44 and valor >= 80.31:
        return 'Classe_C'
    else:
        return 'Classe_D'

# Aplicar a função à coluna existente para criar a nova coluna 'classe'
df['classificacao_hospedes'] = df['average price'].apply(atribuir_classe)

df.head()
```
